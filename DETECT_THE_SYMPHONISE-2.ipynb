```python
import librosa
import numpy as np
from scipy.signal import find_peaks
import pretty_midi

# Charger le fichier audio
audio_path = "/content/Musique2.wav"

y, sr = librosa.load(audio_path)


# Extraction des caractéristiques de la musique
y_harmonic, y_percussive = librosa.effects.hpss(y)
chroma = librosa.feature.chroma_cqt(y=y_harmonic, sr=sr)
chroma = np.mean(chroma, axis=1)

# Détection des notes de musique
f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))
notes = librosa.hz_to_midi(f0)

# Identifier les notes
notes_idx = find_peaks(notes)[0]
note_names = [pretty_midi.note_number_to_name(int(np.round(notes[i]))) for i in notes_idx]
print(note_names)
```

```python
import regex as re
li = []
for element in (note_names):
    li.append(re.sub(r'[#|\d+]', '', element))
```

```python
li
```

```python
tt = []
dos = {'C':'DO', 'D':'RE','E':'MI','F':'FA','G':'SOL','A':'LA','B':'SI'}
for element in li :
  if element in dos.keys():
    element = dos[element]
    tt.append(element)
```

```python
import matplotlib.pyplot as plt
```

```python
import collections

word_freq = collections.Counter(tt)

# Extract the word and frequency data into separate lists
words = list(word_freq.keys())
freqs = list(word_freq.values())

# Create a pie chart
plt.pie(freqs, labels=words, autopct='%1.1f%%')

# Add a title
plt.title('Word Frequencies')

# Show the chart
plt.show()
```
